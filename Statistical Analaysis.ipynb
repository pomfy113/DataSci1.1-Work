```python
import pandas as pd
import seaborn as sns
import scipy.stats

df = pd.read_csv('Iris.csv')
df = df.drop(columns=['Id', 'Species'])
display(df.corr())
sns.heatmap(df.corr(), annot=True)
```

```python
def correlation(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    num = [(x_data - x_mean)*(y_data - y_mean) for x_data, y_data in zip(x,y)]
    # Tried to shorten it to no avail
    den_1 = [(i - x_mean)**2 for i in x]
    den_2 = [(j - y_mean)**2 for j in y]
    
    correlation_x_y = np.sum(num)/np.sqrt(np.sum(den_1))/np.sqrt(np.sum(den_2))
    return correlation_x_y

print(correlation(df['SepalLengthCm'], df['PetalLengthCm']))
print(scipy.stats.pearsonr(df['SepalLengthCm'], df['PetalLengthCm']))
```
