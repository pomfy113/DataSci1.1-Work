```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

filename = './kaggle_income.csv'
income_df = pd.read_csv(filename, encoding='cp1252')
```

```python
income_df.head()
```

```python
income_mean = income_df['Mean']
sb.distplot(income_mean)
```

```python
states = income_df['State_Name']
sb.countplot(y=states)
```

```python
income_median = income_df['Median']
sb.distplot(income_median)
```

```python
sb.barplot(y=income_mean, x=states)
```

```python
data = income_df.sort_values('Mean', ascending=False)[:10]
means = data.Mean
states = data.State_Name
sb.barplot(y=states, x=means)
```
