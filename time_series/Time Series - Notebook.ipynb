```python
data = pd.read_csv('AAPL.csv')
data['Date'] = pd.to_datetime(data['Date'])
data['Tick'] = range(0, len(data))
plt.plot(data['Date'], data['Close'])
plt.show()
```

```python
sample = data.sample(frac=0.5)
sample = sample.sort_values(by=['Tick'])
plt.plot(sample['Date'], sample['Close'])
```

```python
sample['Rolling_Mean'] = data['Close'].rolling(window=5).mean()
```

```python
plt.plot(sample['Date'], sample['Rolling_Mean'])
```
